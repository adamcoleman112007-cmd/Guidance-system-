```python
python3 -m venv yolo_tracker
source yolo_tracker/bin/activate
pip install --upgrade pip

pip install ultralytics opencv-python numpy

cat << 'EOF' > object_tracker.py
import cv2
from ultralytics import YOLO

def main():
    # Load YOLOv8 nano model (fastest, auto-downloads)
    model = YOLO("yolov8n.pt")

    # Open Mac laptop camera
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        raise RuntimeError("ERROR: Could not open camera")

    # Set resolution (good balance for Mac CPU)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

    print("Press ESC to quit")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Run YOLO detection + tracking
        results = model.track(
            frame,
            persist=True,
            conf=0.4,
            iou=0.5,
            tracker="bytetrack.yaml"
        )

        if results and results[0].boxes.id is not None:
            boxes = results[0].boxes.xyxy.cpu().numpy()
            ids = results[0].boxes.id.cpu().numpy()
            classes = results[0].boxes.cls.cpu().numpy()

            for box, track_id, cls in zip(boxes, ids, classes):
                x1, y1, x2, y2 = map(int, box)
                label = model.names[int(cls)]

                # Draw bounding box
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

                # Draw label + tracking ID
                cv2.putText(
                    frame,
                    f"{label} ID:{int(track_id)}",
                    (x1, y1 - 8),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.6,
                    (0, 255, 0),
                    2
                )

                # Draw center point
                cx = (x1 + x2) // 2
                cy = (y1 + y2) // 2
                cv2.circle(frame, (cx, cy), 4, (0, 0, 255), -1)

        cv2.imshow("YOLOv8 Real-Time Object Tracker", frame)

        if cv2.waitKey(1) & 0xFF == 27:
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
EOF

python object_tracker.py

```
